```python
import pandas as pd

file_path = 'your_data.json'  # Replace with the actual path to your JSON file

try:
    df = pd.read_json(file_path)
    print("JSON file loaded successfully into a Pandas DataFrame.")
    # Now you can work with the 'df' DataFrame
    # For example, to see the number of rows (houses):
    number_of_rows = len(df)
    print(f"Number of rows in the dataset: {number_of_rows}")

except FileNotFoundError:
    print(f"Error: The file '{file_path}' was not found.")
except Exception as e:
    print(f"An error occurred while loading the JSON file: {e}")
```

```python
import pandas as pd
from google.colab import files

# Step 1: Upload the file
print("Please upload your Excel file.")
uploaded = files.upload()

# Assuming you uploaded one file, get the filename
# The keys of the 'uploaded' dictionary are the filenames
if uploaded:
    file_name = list(uploaded.keys())[0]
    print(f"File '{file_name}' uploaded successfully.")

    # Use the actual uploaded file name
    file_path = file_name

    try:
        df = pd.read_excel(file_path)

        # Get the number of rows in the DataFrame
        number_of_houses = len(df)

        print(f"The total number of houses in the dataset is: {number_of_houses}")

    except FileNotFoundError:
        # This catch is less likely now that we're using the uploaded name,
        # but kept for robustness.
        print(f"Error: The file '{file_path}' was not found after upload.")
    except Exception as e:
        print(f"An error occurred while reading the Excel file: {e}")
else:
    print("No file was uploaded.")
```
